```python
import pandas as pd
import numpy as np

stu_path = "students_complete.csv"
school_path = "schools_complete.csv"

student_df = pd.read_csv(stu_path)
school_df = pd.read_csv(school_path)
```

```python
total_schools = len(school_df['school_name'])
total_students = len(student_df['student_name'])
total_budget = school_df['budget'].sum()
avg_math = student_df['math_score'].mean()
avg_read = student_df['reading_score'].mean()


passing_math = len(student_df.loc[(student_df['math_score'] >= 70), :])
percent_math = round((passing_math / total_students) * 100, 2)

passing_read = len(student_df.loc[(student_df['reading_score'] >= 70), :])
percent_read = round((passing_read / total_students) * 100, 2)

total_passing = (percent_read + percent_math) / 2
```

```python
district_summary_df = pd.DataFrame([{'Total # of Schools': total_schools, 'Total # of Students': total_students,
                                   'Total Budget': total_budget, 'Average Math Score': avg_math, 'Average Reading Score': avg_read,
                                    '% Passing Math': percent_math, '% Passing Reading': percent_read, 
                                    'Overall Passing Rate': total_passing
                                   }])

#Change format of columns
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.2f}'.format)
district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map('{:.2f}'.format)
district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map('{:.2f}'.format)
district_summary_df['Overall Passing Rate'] = district_summary_df['Overall Passing Rate'].map('{:.2f}'.format)
district_summary_df['Total # of Students'] = district_summary_df['Total # of Students'].map('{:,}'.format)

```

```python
# Organize Columns
district_summary = district_summary_df[['Total # of Schools', 'Total # of Students', 'Total Budget', 'Average Math Score',
                                       'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate'
                                       ]]

school_df_test = school_df.copy()
grade_df = school_df.copy()
```

```python
school_math_per = []
school_read_per = []
school_mpass_count = []
school_rpass_count = []
school_tpass_count = []
budget_per_student = []
ninth_math = []
tenth_math = []
eleventh_math = []
twelfth_math = []
ninth_read = []
tenth_read = []
eleventh_read = []
twelfth_read = []

ninth_stu = []
tenth_stu = []
eleventh_stu = []
twelfth_stu = []

#Loop through each student for each school, then append values for each school to a list to be added to the df
for i, name in enumerate(school_df['school_name']):
    
    school_math = []
    school_read = []
    nm = []
    tem = []
    em = []
    twm = []
    
    nr = []
    ter = []
    er = []
    twr = []

    x=0
    y_math=0
    z_read=0
    ninth_count = 0
    tenth_count = 0
    eleventh_count = 0
    twelfth_count = 0
    
    for stu in student_df['school_name']:
        
        if name == stu:
        
            school_math.append(student_df.iloc[x,6])
            school_read.append(student_df.iloc[x,5])
            
            if student_df.iloc[x,6] >= 70:
            
                y_math = y_math + 1
            
            if student_df.iloc[x,5] >= 70:
                
                z_read = z_read + 1
            
            if student_df.iloc[x,3] == '9th':
                
                nm.append(student_df.iloc[x,6])
                nr.append(student_df.iloc[x,5])
                ninth_count = ninth_count + 1
            
            elif student_df.iloc[x,3] == '10th':
                
                tem.append(student_df.iloc[x,6])
                ter.append(student_df.iloc[x,5])
                tenth_count = tenth_count + 1
            
            elif student_df.iloc[x,3] == '11th':
                
                em.append(student_df.iloc[x,6])
                er.append(student_df.iloc[x,5])
                eleventh_count = eleventh_count + 1
                
            elif student_df.iloc[x,3] == '12th':
                
                twm.append(student_df.iloc[x,6])
                twr.append(student_df.iloc[x,5])
                twelfth_count = twelfth_count + 1
                
        
        x = x + 1
        
    ninth_stu.append(ninth_count)
    tenth_stu.append(tenth_count)
    eleventh_stu.append(eleventh_count)
    twelfth_stu.append(twelfth_count)
    
    ninth_math.append(sum(nm) / ninth_count)
    ninth_read.append(sum(nr) / ninth_count)
    tenth_math.append(sum(tem) / tenth_count)
    tenth_read.append(sum(ter) / tenth_count)
    eleventh_math.append(sum(em) / eleventh_count)
    eleventh_read.append(sum(er) / eleventh_count)
    twelfth_math.append(sum(twm) / twelfth_count)
    twelfth_read.append(sum(twr) / twelfth_count)
    
    
    mpass = (y_math / school_df.iloc[i,3]) * 100
    rpass = (z_read / school_df.iloc[i,3]) * 100
    
    school_mpass_count.append((y_math / school_df.iloc[i,3]) * 100)
    school_rpass_count.append((z_read / school_df.iloc[i,3]) * 100)
    school_math_per.append(sum(school_math) / school_df.iloc[i, 3])
    school_read_per.append(sum(school_read) / school_df.iloc[i, 3])
    school_tpass_count.append((mpass + rpass) / 2)
    budget_per_student.append(school_df.iloc[i, 4] / school_df.iloc[i,3])
    
school_df_test['Budget per Student'] = budget_per_student
school_df_test['Average Math Score'] = school_math_per
school_df_test['Average Reading Score'] = school_read_per    
school_df_test['% Passing Math'] = school_mpass_count
school_df_test['% Passing Reading'] = school_rpass_count
school_df_test['Overall Pass Rate'] = school_tpass_count
```

```python
clean_df = school_df_test.rename(columns={'school_name': 'School', 'type': 'Type of School', 'size': '# of Students',
                                         'budget': 'Budget'
                                         })
clean_df['# of Students'] = clean_df['# of Students'].map('{:,}'.format)
clean_df['Budget'] = clean_df['Budget'].map('${:,.2f}'.format)
clean_df['Budget per Student'] = clean_df['Budget per Student'].map('${:.2f}'.format)
clean_df['Average Math Score'] = clean_df['Average Math Score'].map('{:.2f}'.format)
clean_df['Average Reading Score'] = clean_df['Average Reading Score'].map('{:.2f}'.format)
clean_df['% Passing Math'] = clean_df['% Passing Math'].map('{:.2f}'.format)
clean_df['% Passing Reading'] = clean_df['% Passing Reading'].map('{:.2f}'.format)
clean_df['Overall Pass Rate'] = clean_df['Overall Pass Rate'].map('{:.2f}%'.format)

clean_df
```
