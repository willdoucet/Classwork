```python
import pandas as pd
import pymongo
import json
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.project2_db

lfa = db.labor_force_age
lfg = db.labor_force_gender
un = db.unemployment_education
wage = db.wages
job = db.jobs
```

```python
#LABOR FORCE PARTICIPATION DATA
#age
lf_df = pd.read_csv('data/Civilian Labor Force Participation Rate.csv')
lf1619_df = pd.read_csv('data/Civilian Labor Force Participation Rate 16 to 19 years.csv')
lf2024_df = pd.read_csv('data/Civilian Labor Force Participation Rate 20 to 24 years.csv')
lf2554_df = pd.read_csv('data/Civilian Labor Force Participation Rate 25 to 54 years.csv')
lf55_df = pd.read_csv('data/Civilian Labor Force Participation Rate 55 years and over.csv')
#gender
lf_w_df = pd.read_csv('data/Civilian Labor Force Participation Rate Women.csv')
lf_m_df = pd.read_csv('data/Civilian Labor Force Participation Rate Men.csv')

#COMBINE AGE AND GENDER DATAFRAMES
lf_age_df = lf_df.copy()
lf_age_df['Civilian Labor Force Participation Rate 16 to 19 years'] = lf1619_df.iloc[:,1]
lf_age_df['Civilian Labor Force Participation Rate 20 to 24 years'] = lf2024_df.iloc[:,1]
lf_age_df['Civilian Labor Force Participation Rate 25 to 54 years'] = lf2554_df.iloc[:,1]
lf_age_df['Civilian Labor Force Participation Rate 55 years and over'] = lf55_df.iloc[:,1]

lf_g_df = lf_df.copy()
lf_g_df['Civilian Labor Force Participation Rate Women'] = lf_w_df.iloc[:,1]
lf_g_df['Civilian Labor Force Participation Rate Men'] = lf_m_df.iloc[:,1]
```

```python
#UNEMPLOYMENT DATA
un_df = pd.read_csv('data/Civilian Unemployment Rate.csv')
un_college_df = pd.read_csv("data/Unemployment Rate College Graduates Bachelor's Degree and Higher, 25 years and over.csv")
un_high_df = pd.read_csv('data/Unemployment Rate High School Graduates, No College, 25 years and over.csv')
un_no_df = pd.read_csv('data/Unemployment Rate Less than a High School Diploma, 25 years and over.csv')

c_list = un_college_df["Unemployment Rate College Graduates Bachelor's Degree and Higher, 25 years and over"].tolist()
h_list = un_high_df["Unemployment Rate High School Graduates, No College, 25 years and over"].tolist()
n_list = un_no_df["Unemployment Rate Less than a High School Diploma, 25 years and over"].tolist()

#remove dates from total unemployment so dates are same for all series
print(un_college_df.iloc[0])
print(un_df.loc[un_df['DATE'] == '1/1/1992'])
un_df_current = un_df.iloc[528:]

#join dataframes
un_df_current["Unemployment Rate College Graduates Bachelor's Degree and Higher, 25 years and over"] = c_list
un_df_current["Unemployment Rate High School Graduates, No College, 25 years and over"] = h_list
un_df_current["Unemployment Rate Less than a High School Diploma, 25 years and over"] = n_list
```

```python
#WAGE DATA
wage_salary_df = pd.read_csv('data/Compensation of Employees Wages and Salary Accruals.csv')
wage_benefits_df = pd.read_csv('data/Compensation of employees Supplements to wages and salaries.csv')
ben_list = wage_benefits_df['Compensation of employees Supplements to wages and salaries'].tolist()

wage_salary_df['Compensation of employees Supplements to wages and salaries'] = ben_list
wage_salary_df.head()
```

```python
#JOB VACANCY AND HIRES DATA
job_df_start = pd.read_csv('data/Total Unfilled Job Vacancies for the United States.csv')
hires_df = pd.read_csv('data/Hires Total Private.csv')
jobs_list = job_df_start["Total Unfilled Job Vacancies for the United States"].tolist()
hires_list = hires_df["Hires Total Private"].tolist()

job_df = pd.concat([job_df_start, hires_df["Hires Total Private"][0:-2]], axis=1)
```

```python
lf_age_dict = lf_age_df.to_dict('records')
lf_gender_dict = lf_g_df.to_dict('records')
un_dict = un_df_current.to_dict('records')
wage_dict = wage_salary_df.to_dict('records')
job_dict = job_df.to_dict('records')
```

#ENTER INTO MONGODB

```python
lfa.remove({})
lfa.insert_many(lf_age_dict)
```

```python
lfg.remove({})
lfg.insert_many(lf_gender_dict)
```

```python
un.remove({})
un.insert_many(un_dict)
```

```python
wage.remove({})
wage.insert_many(wage_dict)
```

```python
job.remove({})
job.insert_many(job_dict)
```
