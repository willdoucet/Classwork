```python
#associates = red
#bachelors = green
#masters = blue
education_colors = ['green', 'blue', 'green', 'blue', 'blue', 'blue', 'blue', 'red', 'red', 'red']
```

```python
education_fig = gmaps.figure(center=(32.842224, -117.118739), zoom_level=10)
x = 0
for path in total_path_list:
    
    polygon = gmaps.Polygon(path, stroke_color=education_colors[x], fill_color=education_colors[x])
    
    drawing = gmaps.drawing_layer(features=[polygon], show_controls=False)
    
    education_fig.add_layer(drawing)
    
    x+=1
education_fig
```
